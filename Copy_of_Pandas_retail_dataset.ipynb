```python
#import matplotlib for plotting data


import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,all_data.groupby(['months']).sum()['Sales'])
plt.xticks(months)
plt.ylabel('Sales')
plt.xlabel('Month number')
plt.show()
```
