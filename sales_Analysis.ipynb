### dropping rows of NAN

```python
# rows with not a number ,NAN 
nan_values = all_data[all_data.isnull().any(axis=1)]

print (nan_values)
```

```python
# dropping null values

all_data= all_data.dropna(how = 'all') # will drop all the null values (how = any ) will deletes any row with null value but in our case we are using (how = all)
```

```python
all_data
```

```python
nan_values = all_data[all_data.isnull().any(axis=1)]

print (nan_values)
```

### Augument the column with the additional columns ###

# Adding new month columns

```python
all_data['months']=all_data['Order Date'].str[0:2]
all_data
```

### find 'or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
all_data
```

```python
# adding new column month with numeric data

all_data['months']=all_data['Order Date'].str[0:2]
all_data['months']=all_data['months'].astype('int32')
all_data.head(10)
```

### convert columns to the correct type 




```python
all_data['Quantity Ordered']= pd.to_numeric(all_data['Quantity Ordered']) #make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # to int
all_data.head(5)
```

### operators in pandas ###

```python
#greater than

greater_data = all_data[(all_data['months'] > 6)]
greater_data
```

```python
#greater than

greater_data = all_data[(all_data['Quantity Ordered'] > 2)]
greater_data
```

```python
#isin
isin_data = all_data[(all_data['Quantity Ordered']== 2) | (all_data['Quantity Ordered']== 3  )]
isin_data
```

```python
#notna
notna_data = all_data[(all_data['Price Each'].notna()) ]
notna_data
```

```python
all_data = all_data.iloc[5: 7]
all_data
```

```python
#or
or_data = all_data[(all_data['Quantity Ordered'].isin([all_data]) )]
or_data
```

```python
all_data[(all_data['Price Each'] == 11.95)]
```

### add sales column ###

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head(5)
```

####  What was the best month for sales? How much was earned that month? 

```python
# we can group all data with months

all_data.groupby('months').sum()
```

```python
#import matplotlib for plotting data


import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,all_data.groupby(['months']).sum()['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```
